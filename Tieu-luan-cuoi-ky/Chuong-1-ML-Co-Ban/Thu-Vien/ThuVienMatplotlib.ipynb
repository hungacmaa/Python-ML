# Thư viện matplotlib

import thư viện

```python
import matplotlib.pyplot as plt
```

dữ liệu

```python
X = [1,2,3,4,5,6]
Y = [2,3,4,6,1,8]
```

# Vẽ một biểu đồ đường đơn giản bằng plt.pylot

```python
plt.plot(X, Y)
plt.title("Biểu đồ đường đơn giản") # Đặt tên cho biểu đồ
plt.xlabel("Tên trục x") # Đặt tên cho trục x
plt.ylabel("Tên trục y") # Đặt tên cho trục y
```

# Style cho biểu đồ

```python
from matplotlib import style
style.use("Solarize_Light2")
plt.plot(X, Y)
plt.title("Biểu đồ đường đơn giản") # Đặt tên cho biểu đồ
plt.xlabel("Tên trục x") # Đặt tên cho trục x
plt.ylabel("Tên trục y") # Đặt tên cho trục y
```

## Vẽ biểu đồ lồng nhau


Tập dữ liệu thứ 2

```python
X1 = [1,3,7,9,10,11]
Y1 = [1,4,6,8,10,9]
```

```python
style.use("fivethirtyeight")
plt.plot(X,Y,label='Tập 1',c='r')
plt.plot(X1,Y1,label='Tập 2',c='b')
plt.title('Biểu đồ lồng nhau')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend() # Hiện tên của từng đường có trong biểu đồ
```

## Vẽ nhiều biểu đồ trong 1 figure 

```python
plt.figure(figsize=(20,10))
plt.subplot(1,2,1) # tổng các biểu đồ nằm trên 1 hàng, 2 cột và đây là biểu đồ đầu tiên
plt.plot(X, Y)
plt.title("Biểu đồ đường đơn giản") # Đặt tên cho biểu đồ
plt.xlabel("Tên trục x") # Đặt tên cho trục x
plt.ylabel("Tên trục y") # Đặt tên cho trục y

plt.subplot(1,2,2) # tổng các biểu đồ nằm trên 1 hàng, 2 cột và đây là biểu đồ thứ 2
plt.plot(X1, Y1)
plt.title("Biểu đồ đường đơn giản") # Đặt tên cho biểu đồ
plt.xlabel("Tên trục x") # Đặt tên cho trục x
plt.ylabel("Tên trục y") # Đặt tên cho trục y
```

# Vẽ biểu đồ dạng điểm

```python
plt.plot(X,Y,'bo', label='tập 1')
plt.plot(X1,Y1,'r^', label='tập 2')
plt.title('Biểu đồ lồng nhau')
plt.xlabel('Trục x')
plt.ylabel('Trục y')
plt.legend() # Hiện tên của từng đường có trong biểu đồ
```

# Vẽ biểu đồ phân tán

```python
plt.scatter(X,Y,c='g') #scatter
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```

# Vẽ biểu đồ 3d

```python
from mpl_toolkits import mplot3d
ax = plt.axes(projection='3d')
ax.scatter3D(X,Y)
ax.set_xlabel("Trục x")
ax.set_xlabel("Trục y")
plt.show()
# Lưu ý: nếu muốn xem biểu đồ 3d một cách thuận tiện nhất hãy chạy trên cmd
```

# Biểu đồ cột

## Dữ liệu 

### Code

```python
names = ['Hung', 'Huy', 'Cuong', 'Duc', 'Luc', 'Trang']
diem  = X[:]

plt.bar(names, diem)
plt.title('Biểu đồ cột')
plt.xlabel('Tên')
plt.ylabel('Điểm')
plt.show()
```

# Biểu đồ tròn

```python
names = ['Hung', 'Huy', 'Cuong', 'Duc', 'Luc', 'Trang']
diem  = X[:]
ex = [0,0,0.1,0,0,0] # Phần tách ra khỏi hình tròn
plt.figure(figsize=(20, 10))
plt.pie(diem, explode=ex, labels=names)
plt.axis('equal')
plt.title('Biểu đồ tròn')
plt.legend(title="List of part")
plt.show()
```
